Hyper Parameter Tuning

```python
# importing dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
```

```python
# load the dataset into a variable
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# load this into a dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# Add the target column
data_frame['label'] = breast_cancer_dataset.target

data_frame.head()
```

```python
# Always watch out for cases when the data set  is imbalanced.
data_frame.shape
```

```python
# check for missing values
data_frame.isnull().sum()
```

```python
# check the distributiion of target variable
data_frame['label'].value_counts()
# we can see that this dataset is not so balanced but is managable
```

```python
# Separating the features and target
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
# convert to numpy array
X = X.to_numpy()
Y = Y.to_numpy()
```

#Grid Search CV


GridSearchCV is used for determining the best parameters for our model here

```python
# loading the svc model
model = SVC()
```

```python
# hyperparameters of kernel and 4 types of activation functions
param_grid = {'kernel':['linear','rbf','sigmoid','poly'],'C': [0.1, 1, 5, 10, 20, 50, 100]}
```

```python
# so we search for the best configuration and load resilt into classifier
search_container = GridSearchCV(model,param_grid)
```

```python
# fittinf the data to our model
search_container.fit(X,Y)
```

```python
# Print the results of the search
print(search_container.cv_results_)
```

```python
# best parameters
print(search_container.best_params_)
```

```python
# highest accuracy
print(search_container.best_score_)
```

```python
# load the results into a pandas dataframe
results = pd.DataFrame(search_container.cv_results_)
```

```python
# print the head of the results
results.head()
```

```python
grid_search_result = results[["param_C", "param_kernel", "mean_test_score"]]
grid_search_result
```

From GridSearchCV, the highest accuracy is 96

best parameters = {'C': 100, 'kernel': 'linear'}

#Randomized Search CV

```python
# First, load the svc model
model = SVC()
```

```python
# define hyper parameters
param_grid = {'kernel':['linear','rbf','sigmoid','poly'],'C': [0.1, 1, 5, 10, 20, 50, 100]}
```

```python
# gridsearch
search_container = RandomizedSearchCV(model,param_grid)
```

```python
# fitting the data to our model
search_container.fit(X,Y)
```

```python
# print the classifier CV results
print(search_container.cv_results_)
```

```python
# Print out the best parameters
print(search_container.best_params_)
```

```python
# Highest accuracy
highest_accuracy = search_container.best_score_
print(highest_accuracy)
```

```python
# loading the results to pandas dataframe
results = pd.DataFrame(search_container.cv_results_)
```

```python
results.head()
```

```python
randomized_search_result = results[['param_C','param_kernel','mean_test_score']]
randomized_search_result
```

Highest Accuracy = 96

Best parameters = {'C': 100, 'kernel': 'linear'}
