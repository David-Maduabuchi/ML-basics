Importing Dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing block

```python
# Load the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
# Load the data into a pandas dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

Adding the target column to the data frame.

```python
data_frame['label'] = breast_cancer_dataset.target
```

```python
# Number of rows and columns in the data set
data_frame.shape
```

```python
# getting some information about the data observe the data and ensure that it i
# it has no nonnull values

data_frame.info()
```

```python
# Checking for missing values
data_frame.isnull().sum()
```

```python
# Statistical Measures about the data
data_frame.describe()
```

```python
# Checking the distribution of target variables
data_frame["label"].value_counts()
```

1 -----> Benign

2 -----> Malignant


```python
data_frame.groupby("label").mean()
```

Separating the features and targets.

```python
X = data_frame.drop(columns = "label", axis = 1)
Y = data_frame["label"]
# print(Y)
```

Splitting the data into training and testing data


```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# Training the model while using the training the data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# Accuracy of trainin data

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print("The accuracy on train data is = ", training_data_accuracy)
```

```python
# Accuracy of testing data

X_teest_prediction = model.predict(X_test)
training_data_accuracy = accuracy_score(Y_test, X_teest_prediction)
```

```python
print('The accuracy on test data is = ', training_data_accuracy)
```
